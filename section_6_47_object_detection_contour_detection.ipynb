# Contour detection
## Definition
Contours are defined as simply a *curve joining all the continuous points* (along the boundary), having same color or intensity.

Contours are a very useful tool for shape analysis and object detection and recognition.

OpenCV has a built in countour-finder function that can help us differentiate between internal and external contours.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import skimage as ski
```

```python
img = ski.data.binary_blobs()
img = img.astype(np.uint8)
plt.imshow(img, cmap="gray")
```

```python
type(img)
```

```python
# We'll find the contours of these images
# The methods are cv2.RETR_*
# Only external contours: RETR_EXTERNAL
# The img must be in uint8!
contours, hierarchy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE) # This checks internal and external contours
```

```python
type(contours) # tuple
```

```python
type(hierarchy)  # Regula numpy array
```

```python
external_contours = np.zeros(img.shape) # basic black canvas on which the contours will be drawn

for i in range(len(contours)):
    # External contour
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(external_contours, contours, i, 255, -1)
```

```python
plt.imshow(external_contours, cmap="gray")
```

```python
internal_contours = np.zeros(img.shape)
for i in range(len(contours)):
    # Internal contour
    if hierarchy[0][i][3] != -1:
        cv2.drawContours(internal_contours, contours, i, 255, -1)

plt.imshow(internal_contours, cmap="gray")
```
