```python
import cv2
```

```python
img1 = cv2.imread("Computer-Vision-with-Python/DATA/dog_backpack.png")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread("Computer-Vision-with-Python/DATA/watermark_no_copy.png")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
# 1. BLENDING IMAGES OF THE SAME SIZE
img1 = cv2.resize(img1, (1200,1200))
img2 = cv2.resize(img2, (1200,1200))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
# cv2 overblends with the following formula: src1 * alpha + src2 * beta + gamma
blended = cv2.addWeighted(src1=img1, alpha=0.8, src2=img2, beta=0.2, gamma=0)
```

```python
plt.imshow(blended)
```

```python
# OVERLAY SMALL IMAGE ON TOP OF A LARGER IMAGE (NO BLENDING )
# Numpy reassignment
# Grabbing a junk of the smaller image and replacing the chunks in the larger image
```

```python
img1 = cv2.imread("Computer-Vision-with-Python/DATA/dog_backpack.png")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread("Computer-Vision-with-Python/DATA/watermark_no_copy.png")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2, (600,600))
```

```python
plt.imshow(img2)
```

```python
plt.imshow(img1)
```

```python
large_img = img1
small_img = img2
```

```python
x_offset = 0
y_offset = 0
```

```python
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

```python
small_img.shape
```

```python
large_img[y_offset:y_end,x_offset:x_end] = small_img
```

```python
plt.imshow(large_img)
```
