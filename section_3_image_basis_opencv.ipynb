```python
# Section 3: Image Basics with OpenCV
# 12: Opening Image files in a notebook
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread("Computer-Vision-with-Python/DATA/00-puppy.jpg") 
# If the file does not exist cv2.imread() will return type(img) = NoneType!
```

```python
type(img) # OpenCV directly reads in the array as numpy array
```

```python
img.shape
```

```python
plt.imshow(img)
```

OpenCV and matplotlib have different order of the color channels!

matplotlib --> RGB : RED, GREEN, BLUE

OPENCV --> BGR : BLUE, GREEN, RED

plt.imshow(img)

```python
# convert the color channels with cv2.cvtColor
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
# Read in a grayscale image
img_gray = cv2.imread("Computer-Vision-with-Python/DATA/00-puppy.jpg", cv2.IMREAD_GRAYSCALE) 
# Values are between 0..255
```

```python
plt.imshow(img_gray, cmap="gray")  # Change the colormap to get correct representation
```

```python
plt.imshow(fix_img)
```

```python
fix_img.shape
```

```python
# bring image into a new shape to play around with it
new_img = cv2.resize(fix_img, (1000,400))
# Important! cv2 resize has a swapped coordinate system:
# The size of fix_img is 1300 x 1950  ->  this means that the resize applies 1000 to the 1950 and the 400 to the 1300
plt.imshow(new_img)
```

```python
# Resize by ratio
width_ratio = 0.8
height_ratio = 0.2
```

```python
new_img = cv2.resize(fix_img,(0,0),fix_img,width_ratio,height_ratio)
plt.imshow(new_img)
```

```python
# flip an image
new_img = cv2.flip(fix_img, 0)
plt.imshow(new_img)
```
